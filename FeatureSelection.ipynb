```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('train_2.csv')
```

```python
dataset.head()
```

```python
y_train = dataset[['SalePrice']]
y_train.to_csv('y_train.csv',index=False)
```

```python
x_train = dataset.drop(['Id', 'SalePrice'], axis=1)
```

```python
lasso = Lasso(alpha = 0.005, random_state=0)
```

We will use the Lasso regression. So we created an object, took alpha value keeping in mind that the bigger the alphas value is, the lesser number of features are selected.

Random State is actually the seed value. It can be of any value. But it should be kept same for the train and the test set.

```python
feature_sel_model = SelectFromModel(estimator= lasso)
```

SelectFromModel selects features with nonzero coefficients. It is a meta-transformer for selecting features based on importance weights. lasso regressor is used as the estimator. And lasso penalizes some of the features which have higher weights.

```python
feature_sel_model.fit(x_train, y_train)
```

```python
feature_sel_model.get_support()
```

Here, 
1. True indicates that the feature is important, and should be used.
2. False indicates that the feature is not that important and can be skipped.

## Select Features

```python
selected_features = x_train.columns[(feature_sel_model.get_support())]
```

```python
print('Total features: {}'.format((x_train.shape[1])))
print('No. of selected features: {}'.format(len(selected_features)))
print('Features with coefficient shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

#### Observations-
1. There are 82 total features in the dataset.
2. But out of those features, only 21 features are important and hence selected.
3. There are 61 features which are unimportant and their coeeficient shrank to zero and are skipped.

```python
selected_features
```

```python
x_train = x_train[selected_features]
```
