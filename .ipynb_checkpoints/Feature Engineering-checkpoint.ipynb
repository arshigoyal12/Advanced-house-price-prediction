When we're working with kaggle datasets, we are provided with a separate train data and a separate test data. We usually combine both the datasets and then apply feature engineering. In this case data leakage happens. test set get some infomration about the train set. This improves our accuracy.

But when working on a real world IT project, it doesn't work like that. Real world problem necessarily requires a train test split. And we perform feature engineering for both the train data and test data separately. This will reduce data leakage and overfitting.

Since this is a kaggle dataset, we can simply skip the splitting part.

## Missing Values

### For Categorical features

```python
categorical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']

print('Categorical variables along with their missing value percentage: \n')
for feature in categorical_nan:
    print("{}: {}%".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

#### Replacing nan values with labels

```python
def replace_cat_feature(dataset, features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    #creating a new category called 'Missing'
    
    return data

dataset = replace_cat_feature(dataset, categorical_nan)

dataset[categorical_nan].isnull().sum()
```

Handled all the missing labels(talking specifically about the categorical variables using a new label called 'Missing')

### Numerical Features

```python
numerical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtype != 'O']

print('Numerical variables along with their missing value percentage: \n')
for feature in numerical_nan:
    print("{}: {}%".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

#### Handling missing values in numerical features

```python
for feature in numerical_nan:
    median_value = dataset[feature].median()
    
    # create a new column to capture nan value
    dataset[feature + 'nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature].fillna(median_value, inplace=True)

dataset[numerical_nan].isnull().sum()
```

We simply handled the missing values. We used median as out strategy because of the presence of outliers. But we also created a new feature called 'featurenan' for each feature. Now it represents that whether we had a missin value for a record or not. If a particular record had a missin value, then it will be 1, if not, then 0. It simply tells us whether that record had a missing value in that particular column or not.

## Temporal Features
We saw in the EDA part that, the year variables made no sense on their own. But when subtracted from 'YrSold', the features actually made sense. So we'll try to transform those features similarly.

```python
temporal_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']

for feature in temporal_features[:-1]:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
    
dataset[temporal_features].head()
```

'YrSold' feature gives the year when the house was sold. And other features simply give us the number of years before the 'YrSold'. 

## Numerical(Continuous) Features

We observed that Numerical features are skewed and hence we performed log normal distribution to analyse. Log normal distributions clearlygave a positive correlation.

We saw that, we we performed log normal distribution we only got 5 feature. Those are the feature that do not have a 0 in them. So we will convert those variables in the log normal form.

```python
numerical_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in numerical_features:
    dataset[feature] = np.log(dataset[feature])
```

## Handling Rare Categorical Features

we will remove categories which are present less than 1% of the observations.

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
categorical_features
```

##### Find out percentage of each category for all the categorical variables.


```python
for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

```python
dataset.head(50)
```

#### Observation-
We calculated the percentage of each category in all the categorical variables. And if the percentage of that particular category is less than 1%, that feature is added in the temp_df. And check whether the records of that particular feature exist in the temp_df or not. If they exist, then, update in the main dataset as 'Rare_var'

## 5. Outliers Removal

```python
num_features = [feature for feature in dataset.columns if feature not in categorical_features and feature not in temporal_features]
for feature in num_features:
    plt.figure()  # Create a new figure for each boxplot
    dataset.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

#### Observation- Clearly outliers exist

```python
def remove_outliers_iqr(df, numerical_features):
    df_filtered = df.copy()
    
    for feature in numerical_features:
        q1 = df_filtered[feature].quantile(0.25)
        q3 = df_filtered[feature].quantile(0.75)
        iqr = q3 - q1
        
        lower_fence = q1 - 1.5 * iqr
        upper_fence = q3 + 1.5 * iqr
        
        df_filtered = df_filtered[(df_filtered[feature] >= lower_fence) & (df_filtered[feature] <= upper_fence)]
    
    return df_filtered
```

```python
dataset = remove_outliers_iqr(dataset,num_features)
```

```python
for feature in num_features:
    plt.figure()  # Create a new figure for each boxplot
    dataset.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
